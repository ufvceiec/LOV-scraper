```python
import os
```

```python
# Para renombrar los archivos descargados por el scraper. Tienen que encontrarse en la misma carpeta.
names=[]
i = 0
for item in os.listdir("."):
    head, sep, tail = item.partition('_')
    new_name = head + ".n3"
    if new_name in names:
        i += 1
        new_name = new_name + "(" + i + ")"
    print(new_name)
    names.append(new_name)
    os.rename(item, new_name)
```
